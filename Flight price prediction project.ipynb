```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import StaleElementReferenceException , NoSuchElementException
import re
from selenium.webdriver.common.by import By
import requests
```

```python
driver =webdriver.Chrome("D:\webdriver\chromedriver.exe")
driver.get('https://www.makemytrip.com/flights/')
```

```python
f_name=[]
time.sleep(2)
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//span[@class="boldFont blackText"]')
        for i in na:
            f_name.append(i.text)

    except NoSuchElementException:
        f_name.append('-')
        
f_name      
```

```python
d_tim=[]
time.sleep(2)
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="flexOne timeInfoLeft"]/p[1]')
        for i in na:
            d_tim.append(i.text)

    except NoSuchElementException:
        d_tim.append('-')
        
d_tim 
```

```python
d_pl=[]
time.sleep(2)
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="flexOne timeInfoLeft"]/p[2]')
        for i in na:
            d_pl.append(i.text)

    except NoSuchElementException:
        d_pl.append('-')
        
d_pl
```

```python
times=[]
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="stop-info flexOne"]/p')
        for i in na:
            times.append(i.text)

    except NoSuchElementException:
        times.append('-')
        
times
```

```python
stop=[]
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//p[@ class="flightsLayoverInfo"]')
        for i in na:
            stop.append(i.text)

    except NoSuchElementException:
        stop.append('-')
        
stop        
```

```python
a_time=[]
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="flexOne timeInfoRight"]/p[1]')
        for i in na:
            a_time.append(i.text)

    except NoSuchElementException:
        a_time.append('-')
        
a_time
```

```python
a_pl=[]
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="flexOne timeInfoRight"]/p[2]')
        for i in na:
            a_pl.append(i.text)

    except NoSuchElementException:
        a_pl.append('-')
        
a_pl
```

```python
prices=[]
start=0
end=10
for page in range(start,end):
    try:
        na=driver.find_elements(By.XPATH,'//div[@class="makeFlex column relative splitfare "]/p')
        for i in na:
            prices.append(i.text.replace('',''))

    except NoSuchElementException:
        prices.append('-')
        
prices
```

```python
# Separating Departure Time
data['Dep_hour'] = pd.to_datetime(data.Departure_Time).dt.hour
data['Dep_Minute'] = pd.to_datetime(data.Departure_Time).dt.minute
data.head()
```

```python
# Separating Arrival Time
data['Ari_hour'] = pd.to_datetime(data.Arrival_Time).dt.hour
data['Ari_Minute'] = pd.to_datetime(data.Arrival_Time).dt.minute
data.head()
```

```python
# Separating hours and mins

duration = list(data["Flight_Duration"])
for i in range(len(duration)):
    if len(duration[i].split()) != 3: 
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m" # 0m = 0 minute
    else:
        duration[i] = "0h " + duration[i] # 0h = 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0])) # Extract only hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
data["Duration_hours"] = duration_hours
data["Duration_mins"] = duration_mins
data.drop(['Departure_Time','Arrival_Time','Flight_Duration'], axis=1, inplace=True)
data.head()
```

```python
Airline = data[["Airline_Name"]]
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
data.drop(['Departure_Place','Arrival_Place'], axis=1, inplace=True)
data.head()
```

```python
data.replace({'Non stop':0 ,'1 stop':1 }, inplace =True)
data.head()
```

```python
# adding all the datasets
data_1 =pd.concat([data,Airline], axis=1)
# dropping some columns

data_1.drop(["Airline_Name"] ,axis =1 ,inplace =True)
data_1.head()
```

```python
x =data_1.drop([' Price '], axis =1)
y =data_1[' Price ']
```

```python
sc =StandardScaler()
sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y ,test_size=0.25 ,random_state =25)
```

```python
# KNN
knc =KNeighborsRegressor()
knc.fit(x_train,y_train)
```

```python
y_predict =knc.predict(x_test)
```

```python
metrics.r2_score(y_test,y_predict)
```

```python
print("MSE :",metrics.mean_squared_error(y_test,y_predict))
print("RMSE :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
# Decision Tree
dc =DecisionTreeRegressor()
dc.fit(x_train,y_train)
```

```python
y_predict =dc.predict(x_test)
```

```python
metrics.r2_score(y_test,y_predict)
```

```python
print("MSE :",metrics.mean_squared_error(y_test,y_predict))
print("RMSE :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
# Random Forest
rfc=RandomForestRegressor()
rfc.fit(x_train,y_train)
```

```python
y_predict =rfc.predict(x_test)
```

```python
metrics.r2_score(y_test,y_predict)
```

```python
print("MSE :",metrics.mean_squared_error(y_test,y_predict))
print("RMSE :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
