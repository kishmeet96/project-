```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
file.replace({"Manual":0, "Automatic":1},inplace =True)
file.replace({"Petrol":0, "Diesel":1, "Petrol+cng":2},inplace =True)
```

```python
file.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()

f1 =le.fit_transform(file['Model'])
file['Model'] =f1

f2 =le.fit_transform(file['Kilometer'])
file['Kilometer'] =f2

f3 =le.fit_transform(file['Location'])
file['Location'] =f3

f4 =le.fit_transform(file['Price'])
file['Price'] =f4
file
```

```python
file.info()
```

```python
x=file.drop(['Price'], axis=1)
y =file['Price']
```

```python
sc =StandardScaler()
sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y ,test_size=0.30 ,random_state = 10)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
y_pred =rf.predict(x_test)
```

```python
r_square = metrics.r2_score(y_test,y_pred)
print('r square score:-',r_square)
mean_absolute_error(y_test,y_pred)
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
y_pred =dt.predict(x_test)
```

```python
r_square = metrics.r2_score(y_test,y_pred)
print(r_square)
mean_absolute_error(y_test,y_pred)
```

```python
knn = KNeighborsRegressor()
knn.fit(x_train, y_train)
```

```python
y_pred =knn.predict(x_test)
```

```python
r_square = metrics.r2_score(y_test,y_pred)
print(r_square)
mean_absolute_error(y_test,y_pred)
```
